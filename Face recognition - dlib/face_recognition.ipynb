```python
import cv2
import dlib 
import sys
import numpy as np
```

## 원본 사진 출력

```python
img = cv2.imread('linda.jpg')
cv2.imshow('img', img)
cv2.waitKey(0)
```

## dlib 초기화

```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
```

## 얼굴영역 인식하기

```python
img = cv2.imread('linda.jpg')

# detect faces
faces = detector(img)
face = faces[0]
    
# visualize 
img1 = cv2.rectangle(img, pt1=(face.left(), face.top()), pt2=(face.right(), face.bottom()), color=(255,255,255), thickness=2, lineType=cv2.LINE_AA)
    
cv2.imshow('img', img1)
cv2.waitKey(0)
```

## 얼굴의 랜드마크 출력하기

```python
img = cv2.imread('linda.jpg')

# detect faces
faces = detector(img)
face = faces[0]
     
dlib_shape = predictor(img, face)
shape_2d = np.array([[p.x, p.y] for p in dlib_shape.parts()])

for s in shape_2d:
    cv2.circle(img, center=tuple(s), radius=1, color=(255,255,255), thickness=2, lineType=cv2.LINE_AA)    
    
cv2.imshow('img', img)
cv2.waitKey(0)
```
